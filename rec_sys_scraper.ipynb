```python
# <-- STEP 1 -->
# Create a function to scrape the links of articles:

def scrape_links_leafly():
    
    base_url = "https://www.leafly.com/strains?page="
    link_list = []
    driver.get("https://www.leafly.com/strains?page=1")
    total_pages = driver.find_element_by_class_name("mx-lg").text
    total_pages = int(total_pages.split('of ')[-1])

    for i in range(1, total_pages+1):
        print(i)
        url = base_url + str(i)
        driver.get(url)
        time.sleep(1.18)
        source = driver.find_element_by_class_name("strain-playlist-grid")
        source_list = source.find_elements_by_tag_name('a')
        for link in source_list:
            link_list.append(link.get_attribute('href'))
        print(link_list[-30:])
        print()               

    print('-----------------------------------')
    print('Total of', len(link_list), len(set(link_list)), 'urls to scrape for articles')
    print('-----------------------------------')
    return link_list                  
```
